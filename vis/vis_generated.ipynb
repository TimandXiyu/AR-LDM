```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def display_images_with_caption(folder1, folder2, caption):
    # Collect image file paths from both folders
    images1 = [Image.open(os.path.join(folder1, file)) for file in os.listdir(folder1) if file.endswith(('png', 'jpg', 'jpeg'))][:5]
    images2 = [Image.open(os.path.join(folder2, file)) for file in os.listdir(folder2) if file.endswith(('png', 'jpg', 'jpeg'))][:5]

    # Create a subplot with 2 rows
    fig, axs = plt.subplots(2, 5, figsize=(15, 6))

    # Display images from folder1 in row 1 and folder2 in row 2
    for i in range(5):
        axs[0, i].imshow(images1[i])
        axs[0, i].axis('off')
        axs[1, i].imshow(images2[i])
        axs[1, i].axis('off')
    
    plt.subplots_adjust(wspace=0.1, hspace=0.1)

    # Add a caption with a blue background box
    plt.figtext(0.5, 0.01, caption, ha="center", fontsize=15, bbox={"alpha":0.1, "pad":10})

    plt.show()
    
```

```python

# Example usage
folder1 = '../ckpts/ori_pinkhat/0000'
folder2 = '../ckpts/adapt_output_img/pinkhat_output/0000'
display_images_with_caption(folder1, folder2, '')
```
